```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
spaceship.shape
```

**Check for data types**

```python
spaceship.dtypes
```

**Check for missing values**

```python
spaceship.isnull().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
df = spaceship.dropna()
df.shape
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
df1 = df.copy()
df1['Cabin_initial'] = df['Cabin'].str[0]
```

```python
df1['Cabin_initial'].value_counts()
```

- Drop PassengerId and Name

```python
df1.drop(columns=['PassengerId', 'Name', 'Cabin'], axis=1, inplace=True)
df1.rename(columns={'Cabin_initial': 'Cabin'}, inplace=True)
df1.head(10)
```

- For non-numerical columns, do dummies.

```python
df1['Destination'].value_counts()
```

```python
df1['HomePlanet'].value_counts()
```

```python
df1['Transported'].value_counts()
```

```python
df1.dtypes
```

```python
df2 = df1.copy()

bool_cols = ['CryoSleep', 'VIP', 'Transported']

for col in bool_cols:
    df2[col] = (df2[col] == True).astype(int)

df2.head(10)
```

```python
df2.isnull().sum()
```

```python
df_cat = df2[['HomePlanet', 'Destination', 'Cabin']]
df_cat.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)

ohe.fit(df_cat)
cat_np = ohe.transform(df_cat)
cat_np

cat_df = pd.DataFrame(cat_np, columns=ohe.get_feature_names_out(), index=df_cat.index)
cat_df.head()
```

```python
df_num = df2.drop(columns=df_cat.columns)
df_num.head()
```

```python
df3 = pd.concat([df_num, cat_df], axis=1)
df3.head()
```

**Perform Train Test Split**

```python
features = df3.drop(columns="Transported")
target = df3['Transported']

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)

x_train.head()
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

print(f"The accuracy of the model vs the test is {knn.score(x_test, y_test)*100: .2f}%")
print(f"The accuracy of the model vs the training data is {knn.score(x_train, y_train)*100: .2f}%")
```

- Evaluate your model's performance. Comment it

```python
#Now with min-max scaling:
from sklearn.preprocessing import MinMaxScaler

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)

normalizer = MinMaxScaler()

normalizer.fit(x_train)

x_train_norm = normalizer.transform(x_train)
x_test_norm = normalizer.transform(x_test)

x_train_norm = pd.DataFrame(x_train_norm, columns=x_train.columns , index=x_train.index)
x_train_norm.head()
```

```python
y_train.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_norm, y_train)

print(f"The accuracy of the model vs the test is {knn.score(x_test_norm, y_test)*100: .2f}%")
print(f"The accuracy of the model vs the training data is {knn.score(x_train_norm, y_train)*100: .2f}%")
```
